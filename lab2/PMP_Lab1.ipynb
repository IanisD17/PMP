```python
#1 b)

import random

def simul():

  urna = ['red'] * 3 + ['blue'] * 4 + ['black'] * 2

  zar = random.randint(1, 6)

  if zar in [2, 3, 5]:
    urna.append('black')
  elif zar == 6:
    urna.append('red')
  else:
    urna.append('blue')

  return random.choice(urna)

def red(incercari=100000):
    count = 0
    for i in range(incercari):
        if simul() == 'red':
            count += 1
    return count / incercari

print(red());
```

```python
# 2.2 a)

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

x1 = np.random.poisson(1, 1000)
x2 = np.random.poisson(2, 1000)
x3 = np.random.poisson(5, 1000)
x4 = np.random.poisson(10, 1000)

plt.figure(figsize=(15, 10))

liste_date = [x1, x2, x3, x4, rezultate]
titluri = ["Poisson(1)", "Poisson(2)", "Poisson(5)", "Poisson(10)", "Poisson randomizat"]

for i in range(5):
    plt.subplot(2, 3, i + 1)
    plt.hist(liste_date[i], bins=30, color='skyblue', edgecolor='black')
    plt.title(titluri[i])
    plt.xlabel("Valoare")
    plt.ylabel("Frecventa")

plt.tight_layout()
plt.show()




# 2.2 b) Distributiile cu parametru fix, cum ar fi Poisson(1), Poisson(2), Poisson(5) sau Poisson(10) au o forma clara si destul de ordonata.
# De exemplu, Poisson(1) are majoritatea valorilor aproape de 0 si 1, iar Poisson(10) are valorile mai raspandite, in jur de 10.
# Distributia randomizata, in care alegem la intamplare un lambda este mai amestecata. Are atat valori mici, cat si valori mari
# este mai lata si mai dispersata decat distributiile cu lambda fix.


# In lumea reala, intr-un call center de ex, numarul mediu de apeluri pe ora nu e mereu acelasi. Poate fi mai mic intr-o zi normala, dar creste in zile cand apar probleme tehnice.
# Daca alegem un singur lambda si presupunem ca e mereu fix, nu reusim sa modelam corect ce se intampla de fapt.
# Distributia randomizata e mai realista pentru ca tine cont de aceste schimbari si arata mai bine toate situatiile posibile.
```

```python
# 2.2 c)

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

x1 = np.random.poisson(1, 1000)
x2 = np.random.poisson(2, 1000)
x3 = np.random.poisson(5, 1000)
x4 = np.random.poisson(10, 1000)

valori_lambda = [1, 2, 5, 10]
probabilitati = [0.1, 0.1, 0.5, 0.3]

rezultate = []
for _ in range(1000):
    lam = np.random.choice(valori_lambda, p=probabilitati)
    valoare = np.random.poisson(lam)
    rezultate.append(valoare)

plt.figure(figsize=(15, 10))

liste_date = [x1, x2, x3, x4, rezultate]
titluri = ["Poisson(1)", "Poisson(2)", "Poisson(5)", "Poisson(10)", "Poisson randomizat cu lambda=5 mai prob"]

for i in range(5):
    plt.subplot(2, 3, i + 1)
    plt.hist(liste_date[i], bins=30, color='skyblue', edgecolor='black')
    plt.title(titluri[i])
    plt.xlabel("Valoare")
    plt.ylabel("Frecventa")

plt.tight_layout()
plt.show()

# Majoritatea valorilor se vor grupa în jurul lui 5, avand si varful in jurul lui
```
